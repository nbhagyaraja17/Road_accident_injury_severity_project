```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn import datasets
from sklearn import metrics
from sklearn.model_selection import train_test_split

data = pd.read_csv('Data/2. Dataset_pedestrian_final.xlsx - Data.csv')
print(type(data))
#data['column_name'] = data['column_name'].replace(['old_value'],new_value)

data['Number_of_Vehicles'] = data['Number_of_Vehicles'].replace(['>2'],3) # 1st feature


data['Day_of_Week'] = data['Day_of_Week'].replace(['Weekday'],1)# 2nd feature
data['Day_of_Week'] = data['Day_of_Week'].replace(['Weekend'],2)

data['Road_Type'] = data['Road_Type'].replace(['Single carriageway'],1)# 3rd feature
data['Road_Type'] = data['Road_Type'].replace(['Dual carriageway'],2)
data['Road_Type'] = data['Road_Type'].replace(['Slip road'],3)
data['Road_Type'] = data['Road_Type'].replace(['One way street'],4)
data['Road_Type'] = data['Road_Type'].replace(['Missing'],1)
data['Road_Type'] = data['Road_Type'].replace(['Roundabout'],5)

data['Speed_limit'] = data['Speed_limit'].replace(['20'],1)# 4th feature
data['Speed_limit'] = data['Speed_limit'].replace(['30'],2)
data['Speed_limit'] = data['Speed_limit'].replace(['40'],3)
data['Speed_limit'] = data['Speed_limit'].replace(['â‰¥50'],4)
data['Speed_limit'] = data['Speed_limit'].replace(['Missing'],2)

data['Lighting'] = data['Lighting'].replace(['Daylight'],1)# 5th feature
data['Lighting'] = data['Lighting'].replace(['Darkness - lights lit'],2)
data['Lighting'] = data['Lighting'].replace(['Darkness - lights unlit'],3)
data['Lighting'] = data['Lighting'].replace(['Darkness - lighting unknown'],4)
data['Lighting'] = data['Lighting'].replace(['Darkness - no lighting'],5)

data['Crash_Severity'] = data['Crash_Severity'].replace(['Slight'],1)#output
data['Crash_Severity'] = data['Crash_Severity'].replace(['Serious'],2)
data['Crash_Severity'] = data['Crash_Severity'].replace(['Fatal'],3)

X = data[['Number_of_Vehicles','Day_of_Week','Road_Type','Speed_limit','Lighting']]
y = data['Crash_Severity']

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
X_train = np.array(X_train)
Y_train = np.array(Y_train)
```
